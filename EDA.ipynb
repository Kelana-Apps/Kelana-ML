```python
# Specify the column to check for outliers
col = 'Opening_Time'  # Replace with the desired column

# Calculate Q1, Q3, and IQR
Q1 = kelana_tourism_df[col].quantile(0.25)
Q3 = kelana_tourism_df[col].quantile(0.75)
IQR = Q3 - Q1

# Calculate the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = kelana_tourism_df[(kelana_tourism_df[col] < lower_bound) | (kelana_tourism_df[col] > upper_bound)]
print("Number of outliers:", len(outliers))

# Remove outliers
data_cleaned = kelana_tourism_df[(kelana_tourism_df[col] >= lower_bound) & (kelana_tourism_df[col] <= upper_bound)]

# Display the data after removing outliers
print(data_cleaned.head())
```

```python
# Calculate category frequencies and sort them in ascending order
category_counts = kelana_tourism_df['Category'].value_counts().sort_values(ascending=True)

# Create a blue color palette with fading intensity
colors = sns.color_palette('Blues', n_colors=len(category_counts))

# Create a horizontal bar chart for category distribution in ascending order
ax = category_counts.plot(kind='barh', figsize=(10, 6), color=colors)

# Add title and labels
plt.title('Distribution by Tourism Place Categories')
plt.xlabel('Frequency')
plt.ylabel('Category')
plt.show()
```

```python
# Data preparation (assuming kelana_tourism_df is already defined)
top_cities = kelana_tourism_df['City'].value_counts().head(10)
top_cities_df = kelana_tourism_df[kelana_tourism_df['City'].isin(top_cities.index)]

category_counts_by_city = top_cities_df.groupby(['City', 'Category']).size().unstack(fill_value=0)

# Sort by the total number of tourist attractions per city in ascending order
category_counts_by_city['Total'] = category_counts_by_city.sum(axis=1)
category_counts_by_city = category_counts_by_city.sort_values(by='Total', ascending=True)

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(category_counts_by_city.drop('Total', axis=1), annot=True, fmt='d', cmap='Blues',
            cbar_kws={'label': 'Number of Tourist Attractions'}, linewidths=0.5, annot_kws={"size": 10, 'weight': 'bold'})

# Add additional elements for visualization
plt.title('Category Distribution of Tourist Attractions in the Top 10 Cities', fontsize=14)
plt.xlabel('Tourist Attraction Category', fontsize=12)
plt.ylabel('City', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Data preparation (assuming kelana_tourism_df is already defined)
top_cities = kelana_tourism_df['City'].value_counts().head(10)
top_cities_df = kelana_tourism_df[kelana_tourism_df['City'].isin(top_cities.index)]

category_counts_by_city = top_cities_df.groupby(['City', 'Category']).size().unstack(fill_value=0)

# Sort by the total number of tourist attractions per city in ascending order
category_counts_by_city['Total'] = category_counts_by_city.sum(axis=1)
category_counts_by_city = category_counts_by_city.sort_values(by='Total', ascending=True)

# Plot stacked horizontal bar chart with blue color palette
ax = category_counts_by_city.drop('Total', axis=1).plot(kind='barh', stacked=True, figsize=(12, 6), color=sns.color_palette("Blues", n_colors=len(category_counts_by_city.columns)))

# Calculate the average total number of tourist attractions per city
average_total = category_counts_by_city['Total'].mean()

# Add an average line to the plot
ax.axvline(average_total, color='red', linestyle='--', label=f'Average: {average_total:.2f}')

# Add labels to each category
for i, city in enumerate(category_counts_by_city.index):
    x_pos = 0  # Initial position for label placement
    for category in category_counts_by_city.columns[:-1]:  # Exclude the 'Total' column
        width = category_counts_by_city.loc[city, category]
        x_pos += width  # Add category width to x position
        ax.text(x_pos - width / 2, i, str(width), ha='center', va='center', color='white', fontweight='bold')

# Add additional visualization elements
plt.title('Category Distribution of Tourist Attractions by City')
plt.xlabel('Number of Tourist Attractions')
plt.ylabel('City')
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')

# Display the plot
plt.show()
```

```python
# Sort by the total number of tourist attractions per city in ascending order
category_counts_by_city['Total'] = category_counts_by_city.sum(axis=1)
category_counts_by_city = category_counts_by_city.sort_values(by='Total', ascending=True)

# Calculate the percentage per category in each city
category_percentage_by_city = category_counts_by_city.drop('Total', axis=1).divide(category_counts_by_city['Total'], axis=0) * 100

# Plot stacked horizontal bar chart with percentages and blue color palette
ax = category_percentage_by_city.plot(kind='barh', stacked=True, figsize=(12, 6), color=sns.color_palette("Blues", n_colors=len(category_percentage_by_city.columns)))

# Add percentage labels to each section of the bars
for p in ax.patches:
    width = p.get_width()
    if width > 0:
        ax.text(p.get_x() + p.get_width() / 2, p.get_y() + p.get_height() / 2, f'{width:.1f}%',
                horizontalalignment='center', verticalalignment='center', color='white', fontsize=10)

# Add title and labels
plt.title('Category Distribution of Tourist Attractions in the Top 10 Cities with Percentages')
plt.xlabel('Percentage (%)')
plt.ylabel('City')
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
import nltk

# Make sure stopwords are downloaded
nltk.download('stopwords')

# Load stopwords from nltk
stop_words = set(stopwords.words('indonesian'))

# Determine the cities present in the data
cities = kelana_tourism_df['City'].unique()

# Determine the grid size (e.g., 3 columns and as many rows as needed)
cols = 3
rows = (len(cities) // cols) + (1 if len(cities) % cols != 0 else 0)

# Create figure and axes for the grid
fig, axes = plt.subplots(rows, cols, figsize=(cols*6, rows*6))

# Flatten axes array for easy indexing (if more than one row exists)
axes = axes.flatten()

# Create WordCloud for each city
for i, city in enumerate(cities):
    # Get the tourism descriptions for this city
    city_descriptions = " ".join(kelana_tourism_df[kelana_tourism_df['City'] == city]['Description'].dropna())

    # Create WordCloud while excluding stopwords
    wordcloud = WordCloud(
        width=800,
        height=400,
        background_color='white',
        stopwords=stop_words
    ).generate(city_descriptions)

    # Display the WordCloud on the subplot
    axes[i].imshow(wordcloud, interpolation='bilinear')
    axes[i].axis('off')  # Hide axes
    axes[i].set_title(f'{city}', fontsize=12)

# Adjust the layout to avoid overlap
plt.tight_layout()
plt.show()
```
